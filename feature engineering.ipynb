```python
import pandas as pd
import numpy as np
import matplotlib .pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
dataset =pd.read_csv('train.csv')
dataset.head()
```

```python
X_train,x_test,y_train,y_test=train_test_split(dataset,dataset['SalePrice'],test_size=0.1,random_state=0)
```

```python
x_test.shape,X_train.shape
```

```python
features_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype=='O']
features_nan
for feature in features_nan:
    print('{}:{}%missing values'.format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
def replace_num(dataset,features_nan):
    data=dataset.copy()
    data[features_nan]=data[features_nan].fillna('missing')
    return data
dataset=replace_num(dataset,features_nan)
dataset[features_nan].isnull().sum()
```

```python
dataset.head()
```

```python
numerical_with_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype!='O']
print(numerical_with_nan)
for feature in numerical_with_nan:
    print('{}:{}%missing values'.format(feature,np.around(dataset[feature].isnull().mean(),4)))
```

```python
for feature in numerical_with_nan:
    median_value=dataset[feature].median()
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature]=dataset[feature].fillna(median_value)
dataset[numerical_with_nan].isnull().sum() 
```

```python
dataset.head(30)
```

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature]=dataset['YrSold']-dataset[feature]
```

```python
dataset.head()
```

```python
numerical = ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']
for feature in numerical:
    dataset[feature]=np.log(dataset[feature])
dataset.head()
```

```python
categorical_feature=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
categorical_feature
```

```python
for feature in categorical_feature:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin (temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(20)
dataset.info()
```

```python
for feature in categorical_feature:
    labels= dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels={k:i for i ,k in enumerate(labels,0)}
    dataset[feature]=dataset[feature].map(labels)
```

```python
dataset.head(20)
```

```python
scaling=[feature for feature in dataset.columns if feature  not in ['Id','SalePrice']]
len(scaling)
scaling
```

```python
scaler=MinMaxScaler()
scaler.fit(dataset[scaling])
```

```python
x=scaler.fit_transform(dataset[scaling])
x
```

```python
data =pd.concat([dataset[['Id','SalePrice']].reset_index(drop=True),pd.DataFrame(x,columns=scaling)],axis=1)
data
```

```python
data.to_csv('X_train.csv',index=False)
```
